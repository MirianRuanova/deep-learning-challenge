## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
# The target variable in the model is IS_SUCCESSFUL because ndicates whether the funding provided by Alphabet Soup was used effectively or not.
# It is the binary outcome we want to predict: 1 for successful funding usage and 0 for unsuccessful funding usage.
# The features for your model are the remaining columns in the dataset, excluding the target variable.
# These features will be used to train the model to predict the success of funding applications.
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=["EIN", "NAME"])
application_df
```

```python
# Determine the number of unique values in each colum.
# Determining the number of data points for each unique value in columns that have more than 10 unique values is useful
# for understanding the distribution of data within those columns.
# This information can provide insights into the frequency and representation of different categories or values in the dataset.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_df["APPLICATION_TYPE"].value_counts()
```

```python
# Convert the value count into a dictionary
application_type_counts = application_df["APPLICATION_TYPE"].value_counts().to_dict()
application_type_counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
# The for loop iterates over each application type (atype) in the application_type_counts dictionary.
#if the count of the current application type is less than the "cutoff_value", it appends that application type to the
# application_types_to_replace list
cutoff_value = 527
application_types_to_replace = []
for atype in application_type_counts:
    if application_type_counts[atype] < cutoff_value:
        application_types_to_replace.append(atype)
# Replace in dataframe
# After creating the application_types_to_replace list,
# another for loop is used to replace the application types in the "APPLICATION_TYPE" column of the DataFrame with "Other".
# This ensures that the application types identified in the application_types_to_replace list are replaced accordingly.
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classification_counts = application_df["CLASSIFICATION"].value_counts()
classification_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_counts = classification_counts[classification_counts > 1]
classification_counts
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_counts = application_df["CLASSIFICATION"].value_counts().to_dict()
classifications_counts
cutoff_value_c = 1882
classifications_to_replace = []
for classf in classifications_counts:
    if classifications_counts[classf] < cutoff_value_c:
        classifications_to_replace.append(classf)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df.info()
categorical_columns = ['APPLICATION_TYPE', 'AFFILIATION', 'CLASSIFICATION', 'USE_CASE', 'ORGANIZATION', 'INCOME_AMT', 'SPECIAL_CONSIDERATIONS']


dummy_df = pd.get_dummies(application_df[categorical_columns], columns=categorical_columns)

dummy_df.info()

application_df.drop(columns=categorical_columns, inplace=True)

application_df.info()

application_df = pd.concat([application_df, dummy_df], axis=1)

application_df.head()
```

```python
application_df.info()
```

```python
# Split our preprocessed data into our features and target arrays
y=application_df['IS_SUCCESSFUL']
x = application_df.drop(columns=['IS_SUCCESSFUL'])
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

display(y)


```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
num_input_features=len(X_train.columns)

nn = tf.keras.models.Sequential()

# First hidden layer

nn.add(tf.keras.layers.Dense(units=64, activation='relu', input_dim=num_input_features))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=32, activation='relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
num_epochs=100
batch_size=32
nn.fit(X_train_scaled, y_train, epochs=num_epochs, batch_size=batch_size, validation_data=(X_test_scaled, y_test))
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file

model_filename = "AlphabetSoupCharity.h5"

nn.save(model_filename)
```
